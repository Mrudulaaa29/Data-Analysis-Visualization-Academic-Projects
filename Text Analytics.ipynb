```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import word_tokenize,sent_tokenize
sent="Sachin is the greatest cricketer of India. Virat is caption of Indian Cricket team"
tokens=word_tokenize(sent)
sent_tokens=sent_tokenize(sent)
```

```python
print(tokens)
print(sent_tokens)
```

```python
from nltk.corpus import stopwords 
stop_words=stopwords.words('english')
```

```python
removed_stop_word=[w for w in tokens if w.lower() not in stop_words]
print("After Stop word removal:",removed_stop_word)
```

```python
words = [w.lower() for w in removed_stop_word if w.isalpha()]
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
port_stemmer_output = [stemmer.stem(word) for word in words]
print("Stemming output:", port_stemmer_output)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
lemm_op=[lemmatizer.lemmatize(word) for word in words]
print(lemm_op)
```
