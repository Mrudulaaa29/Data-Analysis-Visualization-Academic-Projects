```python
import pandas as pd
import numpy as np 
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data={
    "name":["Mru","Krishna","Sanchit","Nikhil","Abhijeet","Shruti"],
    "maths":[93,96,np.nan,33,72,86],
    "english":[77,23,86,97,np.nan,75],
    "coding":[97,75,72,28,75,np.nan]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df['maths']=df['maths'].fillna(df['maths'].mean())
```

```python
df.isnull()
```

```python
df['english']=df['english'].fillna(df['english'].mean())
```

```python
df['coding']=df['coding'].fillna(df['coding'].mean())
```

```python
df.isnull()
```

```python
plt.boxplot(df['maths'])
```

```python
plt.show()
```

```python
q1=df['maths'].quantile(0.25)
q3=df['maths'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
upper=q3-1.5*IQR
```

```python
df['maths'] = np.where(df['maths'] > upper, upper,
               np.where(df['maths'] < lower, lower, df['maths']))
```

```python
df
```

```python
df['Math_log'] = np.log(df['maths'])
```

```python
df
```

```python
df['English_normalized'] = (df['english'] - df['english'].min()) / (df['english'].max() - df['english'].min())
```
